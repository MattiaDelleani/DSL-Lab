```python
grouped_df = df.groupby(by="sta")
temp_mean = grouped_df["meantemp"].mean()
temp_std = grouped_df["meantemp"].std()


temp_mean.values[0]
```

```python
cmap = plt.get_cmap("gnuplot")

for i, mu in enumerate(temp_mean):
    sigma = temp_std.iloc[i]
    x= np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma), color = cmap(i))
```

We dont'have to normalize data. And most of them have close meantemp, so the weather conditions could be similar (near 20-30 degree)

```python
global_mean= temp_mean.mean()
global_std = temp_std.mean()

print(f"Mean of global data: {global_mean}\nGlobal std: {global_std}")
x= np.linspace(global_mean - 3*global_std, global_mean + 3*global_std, 100)
plt.plot(x, stats.norm.pdf(x, global_mean, global_std))
```

```python

df["date"] = pd.to_datetime(df["date"])
```

```python
df = df.set_index("date")
df = df[df['sta'] == 22508]['meantemp']
```

```python
mtemps = df
mtemps
```

```python
X = list()
y = list()
W = 3

for i in range(mtemps.size - W):
    X.append(mtemps.iloc[i:i+W].values.T)
    y.append(mtemps.iloc[i+W])
```

```python
X.index
```

```python
X_df = pd.DataFrame(X, columns = ["t0","t1","t2"], index = mtemps.index[:mtemps.size-W] )
y_df = pd.DataFrame(y, index = mtemps.index[:mtemps.size-W])
X_df
```

```python
X_train, y_train = X_df.loc["1940":"1944"], y_df.loc["1940":"1944"]
X_train
```

```python
from datetime import date
```

```python
inital_day = date(1944,12,31) + pd.Timedelta(f"{W} days")
inital_day
```

```python
X_test, y_test = X_df.loc[inital_day:], y_df.loc[inital_day:]
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred = y_pred.reshape(360)

y_hat = pd.DataFrame(y_pred, index = y_test.index)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_hat)
```
