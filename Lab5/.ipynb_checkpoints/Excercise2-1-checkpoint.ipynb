```python
shop = df['shop'].value_counts()
amenity = df['amenity'].value_counts()
public_transport = df['public_transport'].value_counts()
highway = df['highway'].value_counts()
#print(shop)

#shop[:10].plot(kind = "bar") #Is there a way to use .hist() and plot by counted values??
threshold = 0.05
filtered_shop = shop[(shop[:]/shop.sum()) > threshold]
filtered_amenity = amenity[(amenity[:]/amenity.sum()) > threshold]
filtered_public_transport = public_transport[(public_transport[:]/public_transport.sum()) > threshold]
filtered_highway = highway[(highway[:]/highway.sum()) > threshold]
filtered_categories = {}
filtered_categories['shop'] = filtered_shop
filtered_categories['amenity'] = filtered_amenity
filtered_categories['public_transport'] = filtered_public_transport
filtered_categories['highway'] = filtered_highway


for i, value in enumerate(filtered_categories['shop'].index):
    print(value)
```

```python
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)

def plot_category(category, filtered_categories, df):
    type_column = filtered_categories[category]
    cmap = get_cmap(len(type_column))
    img = plt.imread(r'C:\Users\matti\Desktop\Magistrale\Data science lab\Labs\DSL-Lab\Lab5\dataset\POI\New_York_City_Map.PNG')
    fig = plt.figure(figsize = [8,8])
       
    for i, cat_type in enumerate(type_column.index):
        pos = df.loc[df[category] == cat_type, ['@lat', '@lon']]
        
        plt.scatter(pos['@lat'], pos['@lon'], c = cmap(i), label = cat_type, alpha = 0.6, marker = ".")
    
    plt.title(f'Scatter plot for {category}')
    plt.legend()
    plt.show()

        
    
    
```
