```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'..\dataset\2d-synthetic.csv')
df
```

```python
colors = ['blue', 'red']
grouped = df.groupby('label')

plt.scatter(df[df['label'] == 0]['x0'], df[df['label'] == 0]['x1'], color ="red", label = "0", marker = ".")
plt.scatter(df[df['label'] == 1]['x0'], df[df['label'] == 1]['x1'], color ="blue", label = "1", marker = "*")
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(df[['x0', 'x1']], df['label'])
```

## Tree Visualization

```python
from sklearn.tree import export_graphviz
dot_code = export_graphviz(clf, feature_names=['x0','x1'])
#print(dot_code)

img = plt.imread(r"C:\Users\matti\Desktop\Magistrale\Data science lab\Labs\DSL-Lab\Lab6\2-2-Tree.PNG")
plt.figure(figsize = [13,13])
plt.imshow(img)
plt.show()
```


As we can see in the graph the Tree is too depth, so it can distinguish too many classes so it's fully depth. We know that there are only two different classes so it could be a good idea to stop it at step 1 where there are 2 leaf nodes, beacause they are quite homogeneneus 198/67, 192/42. So we can operate with a pre-pruinig, stop before it grows too much, or post-pruining
